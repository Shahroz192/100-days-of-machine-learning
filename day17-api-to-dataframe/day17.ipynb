```python
dfs = []

# Loop through the range of pages
for i in range(1, 400):
    # Use an f-string for better readability
    response = requests.get(f'https://api.themoviedb.org/3/movie/top_rated?api_key=8265bd1679663a7ea12ac168da84d2e8&language=en-US&page={i}')

    # Check if the request was successful (status code 200)
    if response.status_code == 200:
        # Extract relevant data and append to the list
        temp_df = pd.DataFrame(response.json()['results'])[['id', 'title', 'overview', 'release_date', 'popularity', 'vote_average', 'vote_count']]
        dfs.append(temp_df)
    else:
        # Print an error message if the request was not successful
        print(f"Error fetching data from page {i}. Status code: {response.status_code}")

# Concatenate all DataFrames in the list into a single DataFrame
df = pd.concat(dfs, ignore_index=True)
```
